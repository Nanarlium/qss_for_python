## 2. 投票率を求める

```python
table['VEPratio'] = table['total'] / table['VEP'] * 100
table['VAPratio'] = table['total'] / (table['VAP']+table['overseas']) * 100
```

```python
table.head()
```

```python
table.describe()
```

```python
diff = table['VEPratio'].mean() - table['VAPratio'].mean()
diff
```

### 投票率の平均

VEP：48.95％

VAP：45.45％

定義より、VAPのはVEPより大きいため、VEPで算出した投票率はVAPのものに比べて約3.49％高くなる

## 3. ANESに対しての投票率の差を求める
### ANES v.s. VEP
ANESの出した投票率とVEPをもとに算出した投票率の差の平均を求める

```python
table['VEPvsANES'] = table['ANES'] -table['VEPratio']
table['VEPvsANES']
```

```python
table['VEPvsANES'].describe()
```

約16.84％の平均差があり、約8.58%から約22.49%の範囲内で差があることが分かる。

### ANES v.s. VAP

ANESの出した投票率とVAPをもとに算出した投票率の差の平均を求める

```python
table['VAPvsANES'] = table['ANES'] -table['VAPratio']
table['VAPvsANES']
```

```python
table['VAPvsANES'].describe()
```

VAPの場合、約20.33％の平均差があり、約11.06%から約26.17%の範囲内で差があることが分かる

集計方法に偏りがないとするなら、VEPのほうがANESに近くまたVAPよりも差の範囲が狭い

### 中間選挙におけるVEPとANESの比較

この結果より大統領選では中間選挙に比べANES投票率は高くなる傾向が存在する。

## 5. 選挙の年を基準とし、前期後期で比較する。
### テーブルを2つに分割する

```python
first = table[table['year'] < table['year'].mean()]
first
```

```python
second = table[table['year'] >= table['year'].mean()]
second
```

### 上半期、下半期それぞれのVEPとANES投票率の平均差を求める

```python
termDiff = pd.DataFrame({
     'first': [first['ANES'].mean(), 
              first['VEPratio'].mean(), 
              first['VEPvsANES'].mean()],
     'second': [second['ANES'].mean(),
                second['VEPratio'].mean(),
                second['VEPvsANES'].mean()]},
     index = ['ANES', 'VEPratio', 'VEPvsANES']
)

termDiff['diff'] = termDiff['second']-termDiff['first']
termDiff
```

上半期に比べ、ANESは3%の上昇

VEP投票率は約1.03%上昇

VEP投票率とANES投票率は約1.97%の上昇が認められた。

よって、ANES投票率のバイアスは時間が進むとともに増加する傾向が認められた。

## 6. 補正済みVAP投票率と比較
### 補正済みVAPを求める
VAPからfelonsと国籍不保持者を除き、補正済みVAP投票率を求める

```python
table['corrVAP'] = table['VAP'] - (table['felons'] + table['noncit'])
```

```python
temp = table['osvoters'].fillna(0).astype(int)
table['corrTotal'] = table['total'] - temp
table['corrVAPratio'] = table['corrTotal']/table['corrVAP']*100
table['corrVAPvsVAP'] = table['corrVAPratio'] - table['VAPratio']
table['corrVAPvsVEP'] = table['corrVAPratio'] - table['VEPratio']
table['corrVAPvsANES'] = table['ANES'] - table['corrVAPratio']

table.iloc[:,[0,-6,-5,-4,-3,-2,-1]]
```

### 補正済みVAP投票率とその他の投票率の比較

```python
comp = pd.DataFrame({
     'mean': [table['ANES'].mean(),
              table['VEPratio'].mean(),
              table['VAPratio'].mean(),
              table['corrVAPratio'].mean(),   
              table['VEPratio'].mean() - table['VAPratio'].mean(),
              table['VAPvsANES'].mean(),
              table['corrVAPvsVAP'].mean(),
              table['corrVAPvsVEP'].mean(),
              table['corrVAPvsANES'].mean()]},
     index = ['ANES','VEPratio','VAPratio','corrVAPratio', 
              'VAPvsVEP', 'VAPvsANES',
              'corrVAPvsVAP','corrVAPvsVEP','corrVAPvsANES']
 )

comp
```

補正済みVAPでは投票権を持たない人口をカウントしていないため、VAPに比べて投票率が約4.25%上昇している。

またVAPとVEPの補正前平均差が約3.49%あったのに比べ、補正後には約0.76%まで差が縮まっている。

同じくAMES投票率との平均差も補正前の約20.33%から補正後は約16.08%の差へと縮まっている。

これらの結果から、VAPを補正することで補正前の投票率に比べより精度の高い投票率を算出できると考えられる。
